```python
from skimage import io
import matplotlib.pyplot as plt
from skimage.transform import resize




im=io.imread('images/Hoplomyzon_cardosoi-novataxa_2017-Carvalho-Reis-et-Friel.jpg' )
print(im.shape)
io.imshow (im)
plt.show()


```

```python
im=io.imread('images/3D-Micro-CT-of-a-Mandible.png' )
print(im.shape)
io.imshow (im)
plt.show()

```

```python

im_2d=io.imread('images/Hoplomyzon_cardosoi-novataxa_2017-Carvalho-Reis-et-Friel.jpg')
im_3d=io.imread('images/3D-Micro-CT-of-a-Mandible.png')
im_3d_resized=resize(im_3d, (650, 650))
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True,figsize=(12, 5))
#subplot_kw={'adjustable': 'box-forced'}
axes[0].imshow(im_2d,aspect='auto')
axes[0].set_xticks([]), axes[0].set_yticks([])
axes[0].set_title("Hoplomyzon cardosoi novataxa 2017 Carvalho Reis et Friel")
axes[1].imshow(im_3d_resized,aspect='auto')
axes[1].set_xticks([]), axes[1].set_yticks([])
axes[1].set_title("3D-Micro-CT-of-a-Mandible ")
plt.tight_layout()
plt.show()

```
