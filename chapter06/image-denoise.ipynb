```python
import matplotlib.pyplot as plt
from skimage import io
from skimage import restoration
from skimage import img_as_float

im=io.imread('resources/amnig.jpg'  )

im_float = img_as_float(im)
im_denoised = restoration.denoise_nl_means(im_float, h=0.05  )
 

fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True,figsize=(12, 5))
axes[0].imshow(im,aspect='auto')
axes[0].set_xticks([]), axes[0].set_yticks([])
axes[0].set_title("Original image")
axes[1].imshow(im_denoised,aspect='auto')
axes[1].set_xticks([]), axes[1].set_yticks([])
axes[1].set_title("De-noise image by non local mean")
plt.tight_layout()
plt.show()


```
