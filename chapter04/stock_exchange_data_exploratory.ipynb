```python
from pandas.tools.plotting import scatter_matrix
import pandas_datareader.data as web
from pandas import Series, DataFrame
import pandas as pd

from numpy.random import randn
import numpy as np
pd.options.display.max_rows = 12
np.set_printoptions(precision=4, suppress=True)
import matplotlib.pyplot as plt
from pylab import pcolor, show, colorbar, xticks, yticks
plt.rc('figure', figsize=(16, 16))
%matplotlib inline
names = ['AAPL', 'BBRY', '^GSPC', 'AMZN']
def get_px(stock, start, end):
    return web.get_data_yahoo(stock, start, end,interval='m')['Adj Close']
px = DataFrame({n: get_px(n, '1/1/2010', None) for n in names})
px.plot(subplots = True, figsize = (10, 10));
plt.legend(loc = 'best')
plt.show()
```

```python
plt.acorr(px [  'AAPL' ] , marker=None, linestyle='-', color='blue', usevlines=False,maxlags=59)
plt.xlabel('AAPL')
plt.show()
plt.acorr(px [  'AMZN' ] , marker=None, linestyle='-', color='green', usevlines=False,maxlags=59)
plt.xlabel('AMZN')
plt.show()
plt.acorr(px [  '^GSPC' ] , marker=None, linestyle='-', color='red', usevlines=False,maxlags=59)
plt.xlabel('^GSPC')
plt.show()
plt.acorr(px [  'BBRY' ] , marker=None, linestyle='-', color='cyan', usevlines=False,maxlags=59)
plt.xlabel('BBRY')
plt.show()
```

```python
m_return=px / px.shift(1) - 1
m_return=m_return[1:] #remove NAN from first row
scatter_matrix(m_return, alpha=0.2, figsize=(6, 6), diagonal='kde')
m_return.plot(subplots = True, figsize = (10, 10));
```

```python
import seaborn as sns
plt.subplots(figsize=(12, 12))
plt.title('Pearson Correlation of features')
sns.heatmap(px.astype(float).corr(), linewidths=0.25, vmax=1.0, square=True, cmap="cubehelix", linecolor='k',
            annot=True)
```
