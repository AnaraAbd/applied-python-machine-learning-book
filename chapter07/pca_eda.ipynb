```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA



import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline



df = pd.read_csv('resources/USArrests.csv', index_col=0)
df.describe()
```

```python
df.head()
```

```python
X_std_array = StandardScaler().fit_transform(df)


df_std=pd.DataFrame(X_std_array, index=df.index, columns=df.columns)
df_std
```

```python

Z_array=PCA().fit(df_std)

pca_subspace_transformation_vector = pd.DataFrame(Z_array.components_.T, index=df_std.columns, columns=['PC1_LV', 'PC2_LV', 'PC3_LV', 'PC4_LV'])
pca_subspace_transformation_vector
```

```python
pca = PCA()


X_transfromed_array = pca.fit_transform(df_std)

df_transfromed = pd.DataFrame(X_transfromed_array, columns=['PC1', 'PC2', 'PC3', 'PC4'], index=df_std.index)
df_transfromed.head()
```

```python
fig , ax1 = plt.subplots(figsize=(16,12))
ax1.set_xlim(-3.5,3.5)
ax1.set_ylim(-3.5,3.5)
#print states in PCA1 and PCA2 space
for i in df_transfromed.index:
  ax1.annotate(i, (-df_transfromed.PC1.loc[i], -df_transfromed.PC2.loc[i]), ha='center')
# Zero center lines
ax1.hlines(0,-3.5,3.5, linestyles='dotted', colors='grey')
ax1.vlines(0,-3.5,3.5, linestyles='dotted', colors='grey')
ax1.set_xlabel('PC1')
ax1.set_ylabel('PC2')
ax2 = ax1.twinx().twiny()
ax2.set_ylim(-1,1)
ax2.set_xlim(-1,1)
ax2.set_xlabel('PC Subspace transformation Vector', color='red')

a = 1.05
for i in pca_subspace_transformation_vector[['PC1_LV', 'PC2_LV']].index:
 ax2.annotate (i, (-pca_subspace_transformation_vector.PC1_LV.loc[i]*a, -pca_subspace_transformation_vector.PC2_LV.loc[i]*a), color='red')
# Plot vectors
ax2.arrow(0,0,-pca_subspace_transformation_vector.PC1_LV[0], -pca_subspace_transformation_vector.PC2_LV[0],color='red',head_width=0.02)
ax2.arrow(0,0,-pca_subspace_transformation_vector.PC1_LV[1], -pca_subspace_transformation_vector.PC2_LV[1],color='red',head_width=0.02)
ax2.arrow(0,0,-pca_subspace_transformation_vector.PC1_LV[2], -pca_subspace_transformation_vector.PC2_LV[2],color='red',head_width=0.02)
ax2.arrow(0,0,-pca_subspace_transformation_vector.PC1_LV[3], -pca_subspace_transformation_vector.PC2_LV[3],color='red',head_width=0.02)



```

```python
print(pca.explained_variance_)
plt.figure(figsize=(7,5))
plt.plot([1,2,3,4], pca.explained_variance_ratio_.cumsum(), '-o')
plt.ylabel('Cumulative variance of PCA components')
plt.xlabel('Number of Principal Components')

plt.xticks([1,2,3,4])



```
